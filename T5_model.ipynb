```python
from transformers import pipeline

# Load directly from Hub
emotion_classifier = pipeline(
    "text-classification",
    model="abhay2727/emotion_model",
    tokenizer="abhay2727/emotion_model"
)

# Label mapping should match your model's training setup
label_mapping = {
    0: "neutral",
    1: "anger",
    2: "fear",
    3: "surprise",
    4: "happiness",
    5: "sadness",
    6: "disgust"
}

def classify_emotion(text):
    result = emotion_classifier(text)[0]
    return label_mapping[int(result['label'].split("_")[-1])]

# Example usage
print(classify_emotion("Chatgpt is taking iver the job market. I am very concerned about my job, even though I have tea stall outside an IT company."))
```

```python
from accelerate import Accelerator
from transformers import (
    T5ForConditionalGeneration,
    T5Tokenizer,
    get_linear_schedule_with_warmup
)
from torch.utils.data import DataLoader
import torch
from datasets import load_dataset
from torch.optim import AdamW # Import AdamW from torch.optim
```

```python
from datasets import load_dataset, Dataset, DatasetDict
dataset = Dataset.from_pandas(df)
tokenizer = T5Tokenizer.from_pretrained("t5-base")
```

```python
split_dataset = dataset.train_test_split(test_size=0.2, seed=42)
test_valid = split_dataset['test'].train_test_split(test_size=0.5, seed=42)

dataset = DatasetDict({
    'train': split_dataset['train'],
    'validation': test_valid['train'],
    'test': test_valid['test']
})
```

```python
def tokenize_function(examples):
    # Create input text with emotion and context
    inputs = [
        f"emotion: {emotion} context: {context}"
        for emotion, context in zip(examples['emotion'], examples['input_text'])
    ]

    # Tokenize both inputs and targets
    model_inputs = tokenizer(
        inputs,
        max_length=128,
        truncation=True,
        padding="max_length"
    )

    # Tokenize targets (responses)
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(
            examples['response'],
            max_length=128,
            truncation=True,
            padding="max_length"
        )

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
```

```python
tokenized_dataset = dataset.map(tokenize_function, batched=True)
tokenized_dataset.set_format(type='torch', columns=['input_ids', 'attention_mask', 'labels'])
```

```python
train_dataloader = DataLoader(
    tokenized_dataset["train"],
    shuffle=True,
    batch_size=8,
    collate_fn=lambda batch: {
        'input_ids': torch.stack([x['input_ids'] for x in batch]),
        'attention_mask': torch.stack([x['attention_mask'] for x in batch]),
        'labels': torch.stack([x['labels'] for x in batch])
    }
)
```

```python
val_dataloader = DataLoader(
    tokenized_dataset["validation"],
    batch_size=8,
    collate_fn=lambda batch: {
        'input_ids': torch.stack([x['input_ids'] for x in batch]),
        'attention_mask': torch.stack([x['attention_mask'] for x in batch]),
        'labels': torch.stack([x['labels'] for x in batch])
    }
)
```

```python
model = T5ForConditionalGeneration.from_pretrained("t5-base")
optimizer = AdamW(model.parameters(), lr=3e-5, weight_decay=0.01)
scheduler = get_linear_schedule_with_warmup(
    optimizer,
    num_warmup_steps=500,
    num_training_steps=len(train_dataloader)*5
)
```

```python
model, optimizer, train_dataloader, val_dataloader, scheduler = accelerator.prepare(
    model, optimizer, train_dataloader, val_dataloader, scheduler
)
```

```python
for epoch in range(5):
    # Training
    model.train()
    total_loss = 0

    for step, batch in enumerate(train_dataloader):
        with accelerator.accumulate(model):
            outputs = model(
                input_ids=batch['input_ids'],
                attention_mask=batch['attention_mask'],
                labels=batch['labels']
            )

            loss = outputs.loss
            accelerator.backward(loss)

            # Gradient clipping
            accelerator.clip_grad_norm_(model.parameters(), 1.0)

            optimizer.step()
            scheduler.step()
            optimizer.zero_grad()

            total_loss += loss.item()

            if step % 100 == 0:
                accelerator.print(f"Epoch {epoch} | Step {step} | Loss: {loss.item():.4f}")

    # Validation
    model.eval()
    val_loss = 0
    for batch in val_dataloader:
        with torch.no_grad():
            outputs = model(
                input_ids=batch['input_ids'],
                attention_mask=batch['attention_mask'],
                labels=batch['labels']
            )
        val_loss += outputs.loss.item()

    # Generate sample responses
    sample_input = "emotion: sadness context: I lost my job today"
    input_ids = tokenizer.encode(sample_input, return_tensors="pt").to(accelerator.device)
    generated = model.generate(
        input_ids,
        max_length=128,
        num_beams=5,
        temperature=0.7,
        early_stopping=True
    )
    sample_output = tokenizer.decode(generated[0], skip_special_tokens=True)

    accelerator.print(f"\nEpoch {epoch} Summary:")
    accelerator.print(f"Train Loss: {total_loss/len(train_dataloader):.4f}")
    accelerator.print(f"Val Loss: {val_loss/len(val_dataloader):.4f}")
    accelerator.print(f"Sample Response: {sample_output}\n")
```

```python
accelerator.wait_for_everyone()
accelerator.save_state(output_dir="./emotion_response_model")
accelerator.save_model(model, "./emotion_response_model")
```
